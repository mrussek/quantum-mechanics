# Uncovering Momentum Space

Time will only be relevant later, so all wave functions are written as $\Psi(x) = \Psi(x, t)$

Recall

$$
\Psi(x) = \frac{1}{\sqrt{2\pi}}\int{\Phi(k)e^{ikx}\d{k}}
$$

Where

$$
\Phi(k) = \frac{1}{\sqrt{2\pi}}\int{\Psi(x)e^{-ikx}\d{x}}
$$

$\Phi$ has the same information as $\Psi$ essentially

You can either know $\Psi(x)$ everywhere or $\Phi(k)$ for all $k$

What does $\Phi(k)$ mean?

### Delta functions

Consider substituting the definition of $\Phi(k)$ into $\Psi(x)$

$$
\begin{align}
\Psi(x) &= \frac{1}{\sqrt{2\pi}}\int{\Phi(k)e^{ikx}\d{k}}\\
&= \frac{1}{\sqrt{2\pi}}\int{e^{ikx}\left(\frac{1}{\sqrt{2\pi}}\int{\Psi(x')e^{-ikx'}\d{x'}}\right)\d{k}}\\\\
&= \frac{1}{2\pi}\iint{\Psi(x')e^{ik(x-x')}\d{x'}\d{k}}\\\\
&= \int{\d{x'}\Psi(x')\frac{1}{2\pi}\int{\d{k}e^{ik(x-x')}}}\\\\
&= \int{\d{x'}\Psi(x')\delta(x' - x)}\\\\
&= \int{\Psi(x')\delta(x' - x)\d{x'}}
\end{align}
$$

Note that $\delta(x' - x) = \delta(x - x')$, since the $k$ integration can be substituted with $-k$.

$$
\delta(x - x') = \frac{1}{2\pi}\int_{-\infty}^\infty{\d{k}e^{ik(x - x')}}
$$

Other identities:

$$
\begin{align}
\delta(-x) &= \delta(x)\\\\
\delta(ax) &= \frac{1}{|a|}\delta(x)
\end{align}
$$

## Parseval identity

Want to understand momentum space

Ask the following question: What happens to normalization when we are in momentum space

$$
\int_{-\infty}^\infty{\Psi^*(x)\Psi(x)\d{x}} = 1
$$

What does this mean for $\Phi(k)$?

Use $k$ for the first factor, $k'$ for the second factor

$$
\begin{align}
\int_{-\infty}^\infty{\Psi^*(x)\Psi(x)\d{x}} &= \int_{-\infty}^\infty{\left(\frac{1}{\sqrt{2\pi}}\int{\Phi^*(k)e^{-ikx}\d{k}}\right)\left(\frac{1}{\sqrt{2\pi}}\int{\Phi(k')e^{ik'x}\d{k'}}\right)\d{x}}\\\\
&= \frac{1}{2\pi}\int_{-\infty}^\infty{\left(\int{\Phi^*(k)e^{-ikx}\d{k}}\right)\left(\int{\Phi(k')e^{ik'x}\d{k'}}\right)\d{x}}\\\\
&= \frac{1}{2\pi}\int_{-\infty}^\infty{\iint{\Phi^*(k)\Phi(k')e^{ix(k'-k)}\d{k}\d{k'}}\d{x}}\\\\
&= \iint{\Phi^*(k)\Phi(k')\frac{1}{2\pi}\int_{-\infty}^\infty{e^{ix(k'-k)}\d{k}\d{k'}}\d{x}}\\\\
&= \iint{\Phi^*(k)\Phi(k')\delta(k - k')\d{k}\d{k'}}\\\\
&= \int_{-\infty}^\infty{\Phi^*(k)\Phi(k)\d{k}} = 1
\end{align}
$$

### Parseval's theorem

$$
\int{|\Psi(x)|^2\d{x}} = \int{|\Phi(k)|^2\d{x}}
$$

Where $\Phi$ is the Fourier transform of $\Psi$.

Therefore, we have that $\Phi(k)$ maintains normalization in time.

Rewrite with $p$ and not $k$.

$$
p = \hbar k \Rightarrow \frac{\d{p}}{\hbar} = \d{k}
$$

$$
\Phi(k) = \tilde{\Phi}(p)
$$

Rewrite Fourier transform identities
$$
\begin{align}
\Psi(x) &= \frac{1}{\sqrt{2\pi}}\int{\tilde{\Phi}(p)e^{\frac{ipx}{\hbar}}\frac{\d{p}}{\hbar}} =\frac{1}{\hbar\sqrt{2\pi}}\int{\tilde{\Phi}(p)e^{\frac{ipx}{\hbar}}\d{p}}\\\\
\tilde{\Phi}(p) &= \frac{1}{\sqrt{2\pi}}\int{\Psi(x)e^{\frac{ipx}{\hbar}}\d{x}}
\end{align}
$$

For more symmetry $\tilde{\Phi}(p) = \Phi(p)\sqrt{h}$, so we have

$$
\begin{align}
\Psi(x) &= \frac{1}{\sqrt{2\pi\hbar}}\int{\Phi(p)e^{\frac{ipx}{\hbar}}\d{p}}\\\\
\Phi(p) &= \frac{1}{\sqrt{2\pi\hbar}}\int{\Psi(x)e^{\frac{ipx}{\hbar}}\d{x}}
\end{align}
$$

Parseval's now reads

$$
\int{|\Psi(x)|^2\d{x}} = \int{|\Phi(k)|^2\d{k}} = \int{|\tilde{\Phi}(p)|^2\frac{\d{p}}{\hbar}} = \int{|\Phi(p)|^2\d{p}}
$$

### Interpretation of $|\Phi(p)|^2$

We can consider $|\Phi(p)|^2$ to be the probability of finding the particle's momentum within the infinitesimal range $p + \d{p}$.

## Three-dimensional Fourier Transforms

$$
\begin{align}
\Psi(\v{x}) &= \frac{1}{(2\pi\hbar)^{\frac{3}{2}}}\iiint{\Phi(\v{p})e^{i\frac{\v{p}\cdot\v{x}}{\hbar}}\d{^3\v{x}}}\\\\
\Phi(\v{p}) &= \frac{1}{(2\pi\hbar)^{\frac{3}{2}}}\iiint{\Psi^*(\v{p})e^{-i\frac{\v{p}\cdot\v{x}}{\hbar}}\d{^3\v{p}}}
\end{align}
$$

Three-dimensional of $\delta$

$$
\delta^3(\v{x} - \v{x'}) = \frac{1}{(2\pi)^3}\int{e^{i\v{k}\cdot(\v{x}-\v{x'})}\d{^3\v{k}}}
$$

Parseval's identity in 3d

$$
\int{|\Psi(\v{x})|^2\d{^3\v{x}}} = \int{|\Phi(\v{p})|^2\d{^3\v{p}}}
$$

## Expectation Values of Operators

Random variable $Q$ that can take values in the set $Q_1, \dots, Q_n$ with probabilities $p_1, \dots, p_n$

$$
\e{Q} = \sum_{i=1}^n{p_i Q_i}
$$

In a quantum system, we have that $\Psi^*(x, t)\Psi(x, t)\d{x}$ is the probability of finding the particle in $(x, x + \d{x})$.

Define the expected value of $\hat{x}$

$$
\e{\hat{x}} = \int{x\Psi^*(x,t)\Psi(x,t)\d{x}} = \i{\Psi}{\hat{x}|\Psi}
$$

$\e{\hat{x}}$ may depend on time

Similarly, we have that $|\Phi(p)|^2\d{p}$ is the probability of finding a particle with momentum in $(p, p + \d{p})$.

$$
\begin{align}
\e{\hat{p}} &= \int{p\Phi^*(p,t)\Phi(p, t)\d{p}}\\\\
&= \int{
    p
    \frac{1}{\sqrt{2\pi\hbar}}\int{\Psi^*(x') e^{\frac{i{x'}p}{\hbar}} \d{x'} }
    \frac{1}{\sqrt{2\pi\hbar}}\int{\Psi(x)e^{-\frac{ixp}{\hbar}}\d{x}}
    \d{p}
    }\\\\
&= \frac{1}{2\pi\hbar}
    \iint{
        \Psi^*(x')\Psi(x)
        \int{p e^{\frac{ip(x' - x)}{\hbar}}\d{p}}
        \d{x}\d{x'}
        }\\\\
&= \frac{1}{2\pi\hbar}
    \iint{
        \Psi^*(x')\Psi(x)
        \int{\frac{-\hbar}{i}\pdo{x}e^{\frac{ip(x' - x)}{\hbar}}\d{p}}
        \d{x}\d{x'}
        }\\\\
&= \iint{
        \Psi^*(x')\Psi(x)
        \frac{-\hbar}{i}
        \pdo{x}
        {
        \frac{1}{2\pi\hbar}
        \int{e^{\frac{ip(x' - x)}{\hbar}}\d{p}}
        \d{x}\d{x'}
        }
        }
        \\\\
&= \iint{
        \Psi^*(x')\Psi(x)
        \frac{-\hbar}{i}
        \pdo{x}
        {
        \delta(x' - x)
        }
        \d{x}\d{x'}
        }
        \\\\
&= \iint{
        \Psi^*(x')\pd{\Psi}{x}
        \frac{\hbar}{i}
        \delta(x' - x)
        \d{x}\d{x'}
        }
        \\\\
&= \int{
    \Psi^*(x)
    \frac{\hbar}{i}\pdo{x}
    {
    \Psi(x)
    }
    \d{x}
    }\\\\
&= \i{\Psi}{\frac{\hbar}{i}\pdo{x}{}|\Psi}\\\\
&= \i{\Psi}{\hat{p}|\Psi}
\end{align}
$$

In general, for an operator $\hat{Q}$, we define

$$
\e{\hat{Q}} = \i{\Psi}{\hat{Q}|\Psi} = \int{\Psi^*(x)\hat{Q}\Psi(x)\d{x}}
$$

Example $\hat{T} = \frac{\hat{p}^2}{2m}$ is the kinetic operator; Compute it's expected value.

$$
\begin{align}
\e{\hat{T}} &= \int{\Psi^*(x)\hat{T}\Psi(x)\d{x}}\\\\
&= \int{\Psi^*(x)\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}\Psi(x)\d{x}}\tag{1}\\\\
&= \int{\Phi^*(p)\hat{T}\Phi(p)\d{p}}\\\\
&= \int{\Phi^*(p)\frac{p^2}{2m}\Phi(p)\d{p}}\\\\
&= \int{\frac{\hbar^2}{2m}\left|\pd{\Psi}{x}\right|^2\d{x}}\tag{Integration By Parts of (1)}
\end{align}
$$

## Time Dependence of Expectation Values

$$
\begin{align}
\do{t}{\e{Q}} &= \do{t}{\int{\Psi^*(x, t)Q\Psi(x, t)}}\\\\
&= \int{\pd{\Psi^*}{t}Q\Psi + \Psi^* Q\pd{\Psi}{t} \d{x}}\\\\
&= \int{\frac{i}{\hbar}(\hat{H}\Psi)^*Q\Psi - \frac{i}{\hbar}\Psi^*Q\hat{H}\Psi\d{x}}
\end{align}
$$

We then multiply both sides by $i\hbar$

$$
\begin{align}
i\hbar\do{t}{\e{Q}} &= \int{\Psi^*Q\hat{H}\Psi - (\hat{H}\Psi)^*Q\Psi \d{x}}\\\\
&= \int{\Psi^*Q\hat{H}\Psi - \Psi^*Q\hat{H}\Psi \d{x}}\\\\
&= \int{\Psi^*(Q\hat{H} - Q\hat{H})\Psi \d{x}}\\\\
&= \e{[Q, H]}
\end{align}
$$
