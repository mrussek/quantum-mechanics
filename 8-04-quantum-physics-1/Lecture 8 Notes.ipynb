**Review of Expectation Values**

So first, we need to a do a quick review of _expected value_ or _expectation values_. For a random variable $Q$ that takes on countably many values $Q_1,\dots,Q_n$, we define the _expected value_ as
$$
\e{Q} = \sum_{i=1}^N{p_i Q_i}.
$$

If instead, the value $Q$ takes on a continuum of values $x \in X$, with probability densities $p(x)$, then we have

$$
\e{Q} = \int_{X}{xp(x)\d{x}}
$$

Now, recall that we can interpret $\Psi^*(x)\Psi(x)$ as the probability density of the position of a particle. This means that the expected value of position, $\e{x}$, is given by

$$
\e{x} = \int_{-\infty}^\infty{x p(x) \d{x}} = \int_{\infty}^\infty{x \Psi^*(x)\Psi(x)\d{x}}
$$

Similarly, we interpreted the Fourier transform of the wave function $\Phi(p)$ as a probability density for finding the particle with a momentum $p$

$$
P(\text{finding particle with momentum}\in (p, p+dp)) = \Phi^*(p)\Phi(p)\d{p}
$$

So therefore, the expected value of $p$ is

$$
\e{p} = \int_{-\infty}^\infty{p\Phi^*(p)\Phi(p)\d{p}}\tag{1}
$$

However, we have just come up with a formula for $\Phi(p)$ in terms of $\Psi(x)$

$$
\Phi(p) = \frac{1}{\sqrt{2\pi\hbar}}\int_{-\infty}^\infty{\Psi(x)e^{\frac{-ipx}{\hbar}}\d{x}}
$$

What happens when we substitute this formula into $(1)$?

$$
\begin{align}
\e{p} &= \int_{-\infty}^\infty{
    \d{p}
% 
    p
%
    \frac{1}{2\pi\hbar}
    \left(
    \int{\d{x}
        \Psi^*(x)e^{\frac{ipx}{\hbar}}
    }
    \right)
%     
    \left(
    \int{\d{x'}
        \Psi(x')e^{\frac{-ipx'}{\hbar}}
    }
    \right)
    }\\\\\\
&= \frac{1}{2\pi\hbar}
    \int{
        \d{p}
        \int{
            \d{x}
            \int{
                \d{x'}
                p
                \Psi^*(x)\Psi(x')
                e^{\frac{ipx}{\hbar}}
                e^{\frac{-px'}{\hbar}}
            }
        }
    }\\\\\\
&= \int{
       \d{x}\Psi^*(x)
       \int{
           \d{x'}\Psi(x')
          \frac{1}{2\pi\hbar}
           \int{
               \d{p}
               p
               e^{\frac{ipx}{\hbar}}
               e^{\frac{-ipx'}{\hbar}}
           }
       }
   }\\\\\\
&= \int{
       \d{x}\Psi^*(x)
       \int{
           \d{x'}\Psi(x')
          \frac{1}{2\pi\hbar}
           \int{
               \d{p}
               \left(\frac{\hbar}{i} \pdo{x}{}\right)
               e^{\frac{ip}{\hbar}(x - x')}
           }
       }
   }\\\\\\
&= \int{
       \d{x}\Psi^*(x)
       \int{
           \d{x'}\Psi(x')
          \left(\frac{\hbar}{i} \pdo{x}{}\right)
          \frac{1}{2\pi\hbar}
           \int{
               \d{p}
               e^{\frac{ip}{\hbar}(x - x')}
           }
       }
   }\\\\\\
&= \int{
       \d{x}\Psi^*(x)
       \int{
           \d{x'}\Psi(x')
          \left(\frac{\hbar}{i} \pdo{x}{}\right)
          \delta(x' - x)
       }
   }\\\\\\
&= \int{
       \d{x}\Psi^*(x)
       \int{
           \d{x'}\Psi(x')
          \left(\frac{\hbar}{i} \pdo{x}{}\right)
          \delta(x' - x)
       }
   }\\\\\\
&= \int{
       \d{x}\Psi^*(x)
       \left(\frac{\hbar}{i} \pdo{x}{}\right)
       \int{
           \d{x'}\Psi(x')
          \delta(x' - x)
       }
   }\\\\\\
&= \int{
       \d{x}\Psi^*(x)
       \left(\frac{\hbar}{i} \pdo{x}{}\right)
       \Psi(x)
   }\\\\\\
\end{align}
$$

So we have shown that

$$
\e{p} = \int{\d{x}\Psi^*(x)
        \hat{p}
       \Psi(x)}
$$

Where $\hat{p} = \frac{\hbar}{i}\pdo{x}{}$

This motivates a general definition for the expected value of an operator $Q$

$$
\e{Q} = \int{\d{x}\Psi^(x)Q\Psi(x)}
$$

### Example

Compute the expected value of the kinetic operator

$$
\hat{T} = \frac{\hat{p}^2}{2m}
$$

$$
\begin{align}
\e{\hat{T}} &= \int{ \d{x} \Psi^*(x) \hat{T} \Psi(x) }\\\\
&= \int{
       \d{x}
       \Psi^*(x)
       \frac{-\hbar^2}{2m}
       \frac{\partial}{\partial x^2} 
       \Psi(x)
   }
\end{align}
$$

However, if we have the momentum representation of $\Psi$, we can instead compute

$$
\begin{align}
\e{\hat{T}} &= \int{ \d{p} \Phi^*(p) \hat{T} \Phi(x) }\\\\
&= \int{ \d{p} \Phi^*(p) \frac{p^2}{2m} \Phi(p) }
\end{align}
$$

which may be much simpler
