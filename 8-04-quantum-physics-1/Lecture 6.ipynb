## Normalizable wavefunctions and the question of time evolution

In the last lecture, we discussed the interpretation of the wavefunction. In particular, we came up with the so called _Born rule_, which states that the probability of a particle being found in an infinitesimal region $\d{x}$ is given by $\Psi^*(x, t)\Psi(x, t)$. Alternatively, we can say that the probability of finding the particle within the interval $[a, b]$ is given by

$$
P(a \le x \le b) = \int_b^a{\Psi^*(x, t)\Psi(x, t)}\d{x}
$$

From this we can assume that, if we indeed have a one-dimensional particle, that particle must be found _somewhere_. Specifically, we must have that

$$
P(-\infty \le x \le \infty) = \int_{-\infty}^\infty{\Psi^*(x, t)\Psi(x, t)\d{x}} = 1.
\label{eq:normalization} \tag{1}
$$

However, we would hope that this would continue to be true at all times, not at just some particular time. Can we show that if $\eqref{eq:normalization}$ holds for some particular time $t_0$, it must hold for all $t$? In other words, when the wave function evolves according to the Schrödinger equation, is the normalization preserved?

### Boundary conditions

First, we note a simple consequence of $\eqref{eq:normalization}$. In order for the integral to exist, we must for example have that $\Psi(x, t) \to 0$ for $x \to \pm\infty$, since otherwise $|\Psi(x, t)|^2$ will continue to contribute to the integral and thus the integral will diverge.

**Note**: Technically, we do not have to have that $\lim_{x\to\pm\infty} \Psi(x, t) = 0$, since the limit may technically not exist. We do have that $\lim_{x\to\pm\infty} \Psi(x, t) = 0$ if the limit exists at all. However, the situations where this might occur are unphysical and may be ignored, so we may take $\Psi \to 0$ for $x \to \infty$.

$$
\lim_{x\to\pm\infty}{\Psi(x, t)} = 0 \label{eq:zero}\tag{2}
$$

Furthermore, we will also assume that for sufficient $x$,  $\frac{\partial \Psi}{\partial x}$ will be bounded. That is:

$$
, \forall x, |x| > x_0\quad\frac{\partial \Psi}{\partial x} \le C \label{eq:bounded}\tag{3}
$$

**Note**: Earlier, we said that the Schrödinger equation was _linear_, and so any linear combination of solutions was itself a solution. Note that this would imply that if we had a normalized wavefunction $\Psi(x, t)$, we would also have a solution $c\Psi(x, t)$ whose normalization $|c\Psi(x, t)|^2 = c^2|\Psi(x, t)|^2 = c^2 \ne 1$. While this is true, first we note that any wave function that has finite integral $\int_{-\infty}^\infty{|\Psi(x, t)|^2\d{x}} = C < \infty$, we can construct an equivalent wave function $\frac{1}{\sqrt{C}}\Psi(x, t)$ that has normalization 1. We refer to such functions as _normalizable_ wave functions as opposed to normalized wave functions. Note that trivially, normalizable wave functions are closed under finite linear combination. So long as we restrict our solutions to normalizable wave functions, we can simply convert our wave function to a normalized wave function before calculating probabilities with the Born rule.

## Is probability conserved? Hermiticity of the Hamiltonian

Recall our original goal. We wish to show that if $\int_{-\infty}^\infty{|\Psi(x, t_0)|^2\d{x}} = 1$, then $\int_{-\infty}^\infty{|\Psi(x, t)|^2\d{x}} = 1$ for all values of $t$.

First, we define

$$
\begin{align*}
\rho(x, t) &= \Psi^*(x, t)\Psi(x, t) > 0\\\\
N(t) &= \int_{-\infty}^\infty{\rho(x, t)\d{x}}
\end{align*}
$$

Furthermore, by our hypothesis we have that $N(t_0) = 1$. If we can show that 

$$\td{N}{t} = 0,$$

we can show that $N(t)$ must be constant and therefore equal to $1$ at all times $t$.

Expanding,

$$
\begin{align}
\td{N}{t} &= \do{t}\int_{-\infty}^\infty{\rho(x, t)\d{x}}\\\\
&= \int_{-\infty}^\infty{\pd{\rho}{t}}\d{x}\\\\
&= \int_{-\infty}^\infty{\pdo{t}{\left(\Psi^*\Psi\right)}\d{x}}\\\\
&= \int_{-\infty}^\infty{\left(\pd{\Psi^*}{t}\Psi + \Psi^*\pd{\Psi}{t}\right)\d{x}}\\\\
\end{align}
$$

Now recall the Schrödinger equation

$$
i\hbar\pd{\Psi}{t} = \hat{H}\Psi.
$$

Rearranging, we have

$$
\pd{\Psi}{t} = \frac{-i}{\hbar}\hat{H}\Psi.
$$

Taking the complex conjugate, we have

$$
\pd{\Psi^*}{t} = \frac{i}{\hbar}\left(\hat{H}\Psi\right)^*.
$$

Note that we the complex conjugate commutes with differentiation.

Now, substituting into the above, we have

$$
\begin{align}
\td{N}{t} = \int_{-\infty}^\infty{\left(\pd{\Psi^*}{t}\Psi + \Psi^*\pd{\Psi}{t}\right)\d{x}} &= \int_{-\infty}^\infty{\left(\frac{i}{\hbar}\left(\hat{H}\Psi\right)^*\Psi + \Psi^*\frac{-i}{\hbar}\hat{H}\Psi\right)\d{x}}\\\\
&= \frac{i}{\hbar}\int_{-\infty}^\infty{\left(\left(\hat{H}\Psi\right)^*\Psi - \Psi^*\hat{H}\Psi\right)\d{x}}\\\\
&=\frac{i}{\hbar}\int_{-\infty}^\infty{\left(\hat{H}\Psi\right)^*\Psi\d{x}} - \int_{-\infty}^\infty{\Psi^*\hat{H}\Psi\d{x}}
\end{align}
$$

Now, for this to be 0, we must have

$$
\frac{i}{\hbar}\int_{-\infty}^\infty{\left(\hat{H}\Psi\right)^*\Psi\d{x}} = \int_{-\infty}^\infty{\Psi^*\hat{H}\Psi\d{x}}
$$

This would be true for a _Hermitian_ operator.

**Note**: Before definining Hermitian it serves to define the _adjoint_ of an operator. For any linear operator $T$, we define the adjoint of $T$,  as the operator $T^\dagger$ such that

$$
\i{\Phi}{T\Psi} = \i{\Phi T^\dagger}{\Psi}
$$

In the case of the space of square integrable complex functions of one variable, this reads as

$$
\int_{-\infty}^\infty{\Phi^*(x)T\Psi(x)\d{x}} = \int_{-\infty}^\infty{\left(T^\dagger\Phi(x)\right)^*\Psi(x)\d{x}}
$$

A Hermitian operator is one that is equal to it's own adjoint; that is, an operator $T$ is Hermitian iff

$$
\i{\Phi}{T\Psi} = \i{\Phi T}{\Psi}.
$$

Again, in the space of square integrable functions

$$
\int_{-\infty}^\infty{\Phi^*(x)T\Psi(x)\d{x}} = \int_{-\infty}^\infty{\left(T\Phi(x)\right)^*\Psi(x)\d{x}}
$$

Now, we note that if the Hamiltonian $\hat{H}$ is Hermitian, then we can conclude that $\td{N}{t} = 0$, but we pursue the proof in a different way. 

## Probability current and current conservation

In order to show that $\td{N}{t} = 0$, we can simply show that $\pd{\rho}{t} = 0$, since $\td{N}{t} = \int_{-\infty}^\infty{\pd{\rho}{t}\d{x}}$.

Recall that

$$
\begin{align}
\pd{\rho}{t} &= \frac{i}{\hbar}\left(\left(\hat{H}\Psi\right)^*\Psi - \Psi^*\hat{H}\Psi\right)\\\\
&= \frac{i}{\hbar}\left(\left(-\frac{\hbar^2}{2m}\frac{\partial^2 \Psi}{\partial  x^2} + V\Psi\right)^*\Psi - \Psi^*\left(-\frac{\hbar^2}{2m}\frac{\partial^2 \Psi}{\partial  x^2} + V\Psi\right)\right)\\\\
&= \frac{i}{\hbar}\left(\left(-\frac{\hbar^2}{2m}\frac{\partial^2 \Psi^*}{\partial  x^2} + V\Psi^*\right)\Psi - \Psi^*\left(-\frac{\hbar^2}{2m}\frac{\partial^2 \Psi}{\partial  x^2} + V\Psi\right)\right)\\\\
&= \frac{i}{\hbar}\left(-\frac{\hbar^2}{2m}\frac{\partial^2 \Psi^*}{\partial  x^2}\Psi + V\Psi^*\Psi + \frac{\hbar^2}{2m}\Psi^*\frac{\partial^2 \Psi}{\partial  x^2} -  V\Psi^*\Psi\right)\\\\
&= \frac{i}{\hbar}\left(-\frac{\hbar^2}{2m}\frac{\partial^2 \Psi^*}{\partial  x^2}\Psi + \frac{\hbar^2}{2m}\Psi^*\frac{\partial^2 \Psi}{\partial  x^2}\right)\\\\
&= \frac{i\hbar}{2m}\left(\Psi^*\frac{\partial^2 \Psi}{\partial  x^2}-\frac{\partial^2 \Psi^*}{\partial  x^2}\Psi\right)\\\\
\end{align}
$$

Note that

$$
\begin{align}
\do{x}{\left(\pd{\Psi^*}{x}\Psi - \Psi^*\pd{\Psi}{x}\right)} &= \frac{\partial^2 \Psi^*}{\partial x^2}\Psi + \pd{\Psi^*}{x}\pd{\Psi}{x} - \pd{\Psi^*}{x}\pd{\Psi}{x} - \Psi^*\frac{\partial^2 \Psi}{\partial x^2}\\\\
&= \frac{\partial^2 \Psi^*}{\partial x^2}\Psi - \Psi^*\frac{\partial^2 \Psi}{\partial x^2}
\end{align}
$$

So that we have that

$$
\pd{\rho}{t} = \frac{i\hbar}{2m}\do{x}{\left(\pd{\Psi^*}{x}\Psi - \Psi^*\pd{\Psi}{x}\right)}.
$$

Now, note that

$$
\pd{\Psi^*}{x}\Psi - \Psi^*\pd{\Psi}{x} = \pd{\Psi}{x}\Psi^* - \left(\pd{\Psi}{x}\Psi^*\right)^* = 2i\operatorname{Im}{\left(\pd{\Psi}{x}\Psi^*\right)}
$$

so that we can rewrite the equation above as

$$
\begin{align}
\pd{\rho}{t} &= \frac{i\hbar}{2m}\do{x}{2i\operatorname{Im}{\left(\pd{\Psi}{x}\Psi^*\right)}}\\\\
&= -\pdo{x}{\frac{\hbar}{m}\operatorname{Im}{\left(\pd{\Psi}{x}\Psi^*\right)}}
\end{align}
$$

We define

$$
J(x, t) = \frac{\hbar}{m}\operatorname{Im}{\left(\pd{\Psi}{x}\Psi^*\right)}
$$

and then write

$$
\td{\rho}{t} = -\pd{J}{x}
$$

Recalling the units of $\hbar$ and $m$, we can find the units of $J$ as

$$
[J] = \frac{[\hbar]}{ML^2} = \frac{ML^2}{T}\frac{1}{ML^2} = \frac{1}{T}
$$

This allows us to interpret $J$ as a rate of probability flux per unit time.

## Three dimensional current and conservation

In three dimensions, we would define instead

$$
\vec{J}(x, t) = \frac{\hbar}{m}\operatorname{Im}{\Psi^*\nabla\Psi}
$$

And the analog of the above equation would be

$$
\pd{\rho}{t} + \nabla\cdot \vec{J} = 0
$$

Lastly, returning to the proof of $\td{N}{t} = 0$, we can now see that

$$
\begin{align}
\td{N}{t} &= \int_{-\infty}^\infty{\pd{\rho}{t}\d{x}}\\\\
&= -\int_{-\infty}^\infty{\pd{J}{x}\d{x}}\\\\
&= -\left[J(\infty) - J(-\infty)\right]\\\\
&= 0
\end{align}
$$

Since $\lim_{x\to\pm\infty}{\Psi(x, t)} = 0$ and $\pd{\Psi}{t}$ is asymptotically bounded.
