1. **Radiative collapse of a classical atom.**

   In a classical universe, we might try to build a hydrogen atom by placing an electron in a circular orbit around a proton. We know, however, that a non-relativistic, accelerating electron radiates energy at a rate given by the Larmor formula:
   
   $$
   \frac{dE}{dt} = -\frac{2}{3}\frac{e^2a^2}{c^3}
   $$
   
   Here $e$ is the electron charge and $a$ is the magnitude of the electron acceleration. So the classical atom may have a stability problem. We want to figure out how big is this effect. In the units we are working the electron potential energy in the presence of the proton is $V = \frac{−e^2}{r}$ and the magnitude of the force of attraction is $e^2/r^2$.
   
   a) Show that for a non-relativistic electron the energy $\Delta E$ lost per revolution is small compared to the electron’s kinetic energy $K$. Do this by computing the ratio $\Delta E / K$. Hence, it is possible to regard the orbit as circular at any instant, even though the electron eventually spirals into the proton.
   
      The energy lost by the photon per revolution is given by:
      
      $$
      \Delta E = \int_0^T{\frac{dE}{dt}dt}
      $$
      
      $$
      \begin{align}
      x(t) &= r\cos{\omega t}\\
      y(t) &= r\sin{\omega t}\\
      \Rightarrow 
      \ddot{x}(t) &= -r\omega^2\cos{\omega t}\\
      \ddot{y}(t) &= -r\omega^2\sin{\omega t}\\
      \Rightarrow v &= r\omega\\
      \Rightarrow a &= r\omega^2 = \frac{4r\pi^2}{T^2}\\
      \Rightarrow T &= \sqrt{\frac{4r\pi^2}{a}}
      \end{align}
      $$
      
      $$
      \begin{align*}
      \Delta E &= \int_0^T{\frac{dE}{dt}dt}\\
      &= \int_0^T{-\frac{2}{3}\frac{e^2a^2}{c^3}dt}\\
      &= -\frac{2}{3}\frac{e^2a^2}{c^3}T\\
      &= -\sqrt{\frac{16e^4a^3\pi^2r}{9c^6}}
      \end{align*}
      $$
      
      $$
      \begin{align*}
      \frac{\Delta E}{K} &= \frac{-\sqrt{\frac{16e^4a^3\pi^2r}{9c^6}}}{\frac{1}{2}mv^2}\\
      &= \frac{-2\sqrt{\frac{16e^4a^3\pi^2r}{9c^6}}}{mr^2\omega^2}\\
      &= \frac{-2\sqrt{\frac{16e^4r^3\omega^6\pi^2r}{9c^6}}}{mr^2\omega^2}\\
      &= -\frac{8e^2\omega\pi}{3c^3}\\
      \end{align*}
      $$
      
      Which is very small as it is proportional $\frac{1}{c^3}$ _and_ $e^2$
