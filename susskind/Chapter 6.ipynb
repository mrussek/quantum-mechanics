# 6.1 Mathematical Interlude: Tensor Products

# 6.1.1 Meet Alice and Bob

Consider two distinct quantum systems possesed by Alice and Bob respectively. Denote Alice's system's state space as $S_A$ and Bob's system's state space as as $S_B$. Let $S_A$ be a two-dimensional spin system, so that $S_A$ is spanned by $\aket{H}, \aket{T}$ while $S_B$ will be a six-dimensional vector space spanned by $\ket{i}, 1 \leq i \leq 6$. Any quantum state of Alice's can be represented as a superposition of the form

$$
\aket{\Psi_A} = \alpha_1\aket{H} + \alpha_2\aket{T}
$$

while any state of Bob's can be represented as

$$
\ket{\Psi_B} = \alpha_1\ket{1} + \alpha_2\ket{2} + \alpha_3\ket{3} + \alpha_4\ket{4} + \alpha_5\ket{5} + \alpha_6\ket{6}
$$

# 6.1.2 Representing the Combined System

_Question:_ How do we represent the state space of both Alice's _and_ Bob's system?

We form the _tensor product_ $S_{AB} = S_A \otimes S_B$ of their state spaces!

The tensor product can be thought of the space spanned by the cartesian product of the bases of $S_A$ and $S_B$, which is 

$$
\{ \aket H, \aket T \} \times \{ \ket 1, \ket 2, \ket 3, \ket 4, \ket 5, \ket 6\}
$$

which we may choose to write as

|$H$|$T$|
|---|---|
|$\ket{H1}$|$\ket{T1}$|
|$\ket{H2}$|$\ket{T2}$|
|$\ket{H3}$|$\ket{T3}$|
|$\ket{H4}$|$\ket{T4}$|
|$\ket{H5}$|$\ket{T5}$|
|$\ket{H6}$|$\ket{T6}$|

We will refer to an arbitrary basis vector as $\ket{ab}$ where $a$ will refer to Alice's component and $b$ will refer to Bob's component.

Thus, any state of the combined system can be written as a sum over each of these basis vectors

$$
\ket{\Psi} = \sum_{a,b}\alpha_{a,b}\ket{ab}
$$

**Note:** States such as $\ket{H1}$ and so forth are _single_ basis vector states.

Additionally, we will assume that Alice's space has a set of observables (Hermitian operators) $\sigma$, whereas Bob's space has observables $\tau$.

It's worth noting that that $\dim S_{AB} = (\dim S_A)(\dim S_B)$, and that we can take the tensor product of an arbitrary number of vector spaces.

$$
S_A \otimes S_B \otimes S_C \otimes \cdots
$$

# 6.2 Classical Correlation

Consider the following thought experiment. Charlie has both a penny and a dime. He gives one to Alice and one to Bob, and neither Alice nor Bob observe their coin until they have separated by several light years. At a set moment in time, Alice looks at her coin, and can instantaneously know what coin Bob has. Consider the case that a penny corresponds to $\sigma = 1$ while a dime corresponds to $\sigma = -1$. Then we have observables for each of Alice and Bob $\sigma_A$ and $\sigma_B$ respectively. We can compute that the expected value of each is $\exp{\sigma_A} = \exp{\sigma_B} = 0$ since the coins are chosen with equal probability of being assigned to Alice and Bob respectively. However, we can note that the expectation of the quantitiy $\sigma_A \sigma_B$ is $-1$ and in fact it will always be the case that $\sigma_A \sigma_B = -1$. Thus, we have the following identity:

$$
\exp{\sigma_A\sigma_B} \neq \exp{\sigma_A}\exp{\sigma_B}
$$

We can define the _correlation_ between $\sigma_A$ and $\sigma_B$ as the size of this difference.

$$
\cor{\sigma_A}{\sigma_B} = \exp{\sigma_A\sigma_B} - \exp{\sigma_A}\exp{\sigma_B} \neq 0
$$

If we have a joint distribution $P(a, b) $over two random variables, then they are independent iff $P(a, b) = P_A(a)P_B(b)$. Furthermore, if two variables are independent they are guaranteed to be uncorrelated $\cor{A}{B} = 0$

**Exercise 6.1:** Prove that if $P(a, b)$ factorizes, then the correlation between $a$ and $b$ is zero.

$$
\begin{align*}
\cor{a}{b} &= \exp{ab} - \exp{a}\exp{b}\\
&= \sum_{a,b}{abP(a, b)} - \sum_a{a P_A(a)}\sum_b{b P_B(b)}\\
&= \sum_{a,b}{abP(a, b)} - \sum_a\sum_b{abP_A(a)P_B(b)}\\
&= \sum_{a,b}{abP_A(a)P_B(b)} - \sum_{a,b}{abP_A(a)P_B(b)}
&= 0
\end{align*}
$$

Takeaways:

1. In classical physics, probabilities express uncertainty about the current state. Everything is in principle knowable, but we are simply uncertain
2. In classical physics, if we know the state of the whole, we must know the state of each component.

Neither of these two statements will hold for quantum mechanics.

# 6.3 Combining Quantum Systems

Consider a linear operator operating on the space $S_{AB}$ spanned by $\ket{ab}$. We can represent this operator in the $\ket{ab}$ basis as a matrix:

$$
M_{a'b',ab} = \bra{a'b'}M\ket{ab}
$$

Here, $M$ will be a $N = (\dim{S_A})(\dim{S_B})$ by $N$ square matrix.

Recall that the vectors $\ket{ab}$ form an orthonormal set:

$$
\i{a'b'}{ab} = \delta_{a'b',ab} = \delta_{aa'}\delta{bb'} = \begin{cases} 1 \quad \mathrm{if}\;a = a', b = b' \\ 0 \quad \mathrm{otherwise} \end{cases}
$$

# 6.4 Two Spins

Let's consider the case where both Alice's and Bob's system are spin systems of dimension $2$. Then, we have as basis vectors for Alice's state space $\aket{u}$ and $\aket{d}$ while Bob's basis vectors are $\ket{u}, \ket{d}$.

Similarly, we will have the standard spin operators for Alice's state space:

$$
\sigma_x, \sigma_y, \sigma_z
$$

and for Bob's state space:

$$
\tau_x, \tau_y, \tau_z
$$

The vectors of the combined state space $S_{AB} = S_A \otimes S_B$ will be:

$$
\ket{uu}, \ket{ud}, \ket{du}, \ket{dd}
$$

# 6.5 Product States

While in general, a state in $S_{AB}$ can be a linear superposition of all of the above basis states, there are some states that will make "intuitive" sense as composite states of the system. These are known as _product states_, and they can be thought of as a pair of independent spin states.

Consider a generic Alice state:

$$
\alpha_u\aket{u} + \alpha_d\aket{d}
$$

and a generic Bob state:

$$
\beta_u\ket{u} + \beta_d\ket{d}
$$

Since these are each states of a spin system, we can infer that

$$
\begin{align*}\tag{6.4}
\alpha_u^*\alpha_u + \alpha_d^*\alpha_d &= 1 \\\\
\beta_u^*\beta_u + \beta_d^*\beta_d &= 1 
\end{align*}
$$

We can form the _product state_ from these two states (from different vectors spaces!) using the following:

$$
\begin{align*}
\ket{product\; state} &= \bigg\{ \alpha_u\aket{u} + \alpha_d\aket{d}\bigg\}\otimes\bigg\{ \beta_u\ket{u} + \beta_d\ket{d}\bigg \}\\\\
&= \alpha_u\beta_u\ket{uu} + \alpha_u\beta_d\ket{ud} + \alpha_d\beta_u\ket{du} + \alpha_d\beta_d\ket{dd}
\end{align*}
$$

> The main feature of a product state is that each subsystem behaves indpendently of the other. If Bob does an experiment on his own subsystem, the result is exactly the same as it would be if Alice's subsytem did not exist. The same is true for Alice, of course.

**Exercise 6.2:** Show that if the two normalization condiations of Eqs. 6.4 are satisfied , then the sate-vector of Eq. 6.5 is automatically normalized as well. In other words, show that for this product state, normalizing the overall state-vector does not put any additional constraints on the $\alpha$'s and $\beta$'s.

$$
\begin{align*}
(\alpha_u\beta_u)^*(\alpha_u\beta_u) &+ (\alpha_u\beta_d)^*(\alpha_u\beta_d) + (\alpha_d\beta_u)^*(\alpha_d\beta_u) + (\alpha_d\beta_d)^*(\alpha_d\beta_d)\\\\
&= \alpha_u^*\beta_u^*\alpha_u\beta_u + \alpha_u^*\beta_d^*\alpha_u\beta_d + \alpha_d^*\beta_u^*\alpha_d\beta_u + \alpha_d^*\beta_d^*\alpha_d\beta_d\\\\
&= \alpha_u^*\alpha_u(\beta_u^*\beta_u + \beta_d^*\beta_d) + \alpha_d^*\alpha_d(\beta_u^*\beta_u + \beta_d^*\beta_d)\\\\
&= \alpha_u^*\alpha_u + \alpha_d^*\alpha_d\\\\
&= 1
\end{align*}
$$

Note that tensor products (a.k.a _tensor product spaces_, _product spaces_) are distinct from product _states_. Tensor products are a whole vector space, while a product state is simply a particular kind of vector in that vector space.

$$
\ket{\psi} = \alpha_{uu}\ket{uu} + \alpha_{ud}\ket{ud} + \alpha_{du}\ket{du} + \alpha_{dd}\ket{dd}
$$

# 6.6 Counting Parameters for the Product State

Each product state has 4 parameters, since each component state introduces 4 parameters (2 per each complex number), but we assume that each state is normalized, reducing the parameters by 1 each, and we assume that the overall phase factor is irrelevant, so we arrive at $8 - 2 - 2 = 4$ parameters.

# 6.7 Entangled States

However, we will note that in general, not all vectors of the product space are product states. The most general state can be represented by:

$$
\ket{\Psi} = \psi_{uu}\ket{uu} + \psi_{ud}\ket{ud} + \psi_{du}\ket{du} + \psi_{dd}\ket{dd}
$$

Note that we have $4\cdot2 = 8$ components, but that the normalization condition removes one of them, and we only have one phase factor that we can ignore. Therefore, the general product space has $6$ degrees of freedom, two more than the space of product states.

The new states can be called the _entangled_ states. A canonical example of an entangled state is the _singlet_ state:

$$
\ket{sing} = \frac{1}{\sqrt 2}(\ket{ud} - \ket{du})
$$

which is a superposition of the $\ket{ud}$ state and the $\ket{du}$ state, and cannot be written as a product state.

In addition to the singlet state, we have the so-called triplet states:

$$
\begin{align*}
\frac{1}{\sqrt 2}(\ket{ud} + \ket{du})\\\\
\frac{1}{\sqrt 2}(\ket{uu} + \ket{dd})\\\\
\frac{1}{\sqrt 2}(\ket{uu} - \ket{dd})\\\\
\end{align*}
$$

which are all, like the singlet state, maximally entangled.

**Exercise 6.3:** Prove that the state $\ket{sing}$ cannot be written as a product state.

A generic product state can be written as:

$$
\alpha_u\beta_u\ket{uu} + \alpha_u\beta_d\ket{ud} + \alpha_d\beta_u\ket{du} + \alpha_d\beta_d\ket{dd}.
$$

Matching with the description of a product state, we can see that

$$
\begin{align}
\alpha_u\beta_u &= 0 \\\\
\alpha_u\beta_d &= \frac{1}{\sqrt 2} \\\\
\alpha_d\beta_u &= -\frac{1}{\sqrt 2} \\\\
\alpha_d\beta_d &= 0
\end{align}
$$

And so
$$
\begin{align*}
\alpha_u^*\alpha_u\beta_u^*\beta_u + \alpha_u^*\alpha_u\beta_d^*\beta_d = \frac{1}{2}\\\\
\alpha_u^*\alpha_u\beta_u^*\beta_u + \alpha_d^*\alpha_d\beta_u^*\beta_u = \frac{1}{2}
\end{align*}
$$

Assume that the singlet state is a product state. Then we have that:

$$
\begin{align*}
&\alpha_u^*\alpha_u\beta_u^*\beta_u + \alpha_u^*\alpha_u\beta_d^*\beta_d\\\\
&= \alpha_u^*\alpha_u(\beta_u^*\beta_u + \beta_d^*\beta_d)\\\\
&= \alpha_u^*\alpha_u = \frac{1}{2} = \alpha_d^*\alpha_d
\end{align*}
$$

which in turn implies
$$
\alpha_u^*\alpha_u \beta_d^*\beta_d = \frac{1}{2} \Rightarrow \beta_d^*\beta_d = 1\\
\alpha_d^*\alpha_d \beta_u^*\beta_u = \frac{1}{2} \Rightarrow \beta_u^*\beta_u = 1
$$

which means that $\beta_u^*\beta_u + \beta_d^*\beta_d = 2 \neq 1$ and so, by contradiction, the singlet state must not be a product state.

> What is it about maximally entangeld states that is so fascinating? I can sum this up in two statements:
> * An entangled state is a complete description of the combined system. No more can be known about it.
> * In a maximally entangled state, nothing is known about the individual subsystems.

# 6.8 Alice and Bob's Observables

Recall Alice's and Bob's observables $\sigma$ and $\tau$. Since Alice and Bob's systems are spin systems, we are aware of three observables for each:

$$
\sigma_x, \tau_x\\
\sigma_y, \tau_y\\
\sigma_z, \tau_z
$$

which corresponde to the x, y, z components of the spin respectively.

How can we extend these operators to act on product space vectors $\ket{ab}$?

Simply by letting each of Alice's operators act on Alice's "half" of the state, and each of Bob's operators act on his "half" of the state, as below (also part of the answer to **Exercise 6.4**):

$$
\begin{align}
\sigma_z \ket{uu} &= \ket{uu}\\
\sigma_z \ket{ud} &= \ket{ud}\\
\sigma_z \ket{du} &= -\ket{du}\\
\sigma_z \ket{dd} &= -\ket{dd}\\
\\
\sigma_x \ket{uu} &= \ket{du}\\
\sigma_x \ket{ud} &= \ket{dd}\\
\sigma_x \ket{du} &= \ket{uu}\\
\sigma_x \ket{dd} &= \ket{ud}\\
\\
\sigma_y \ket{uu} &= i\ket{du}\\
\sigma_y \ket{ud} &= i\ket{dd}\\
\sigma_y \ket{du} &= -i\ket{uu}\\
\sigma_y \ket{dd} &= -i\ket{uu}\\
\\
\tau_z \ket{uu} &= \ket{uu}\\
\tau_z \ket{ud} &= -\ket{ud}\\
\tau_z \ket{du} &= \ket{du}\\
\tau_z \ket{dd} &= -\ket{dd}\\
\\
\tau_x \ket{uu} &= \ket{ud}\\
\tau_x \ket{ud} &= \ket{uu}\\
\tau_x \ket{du} &= \ket{dd}\\
\tau_x \ket{dd} &= \ket{du}\\
\\
\tau_y \ket{uu} &= i\ket{ud}\\
\tau_y \ket{ud} &= -i\ket{uu}\\
\tau_y \ket{du} &= i\ket{dd}\\
\tau_y \ket{dd} &= -i\ket{du}\\
\end{align}
$$
