# Chapter 1: Introduction
## 1.1 Quantum Mechanics is Different

In classical mechanics, the state fully determines the measurements that you can make, and measurements do not affect the state of the system. However in quantum mechanics, the state of the system is _necessarily_ perturbed by any measurement.

## 1.2 Spins and Qubits
The _state_ of an electron is _not_ fully determined by its position, nor its momentum. An additional degree of freedom in the state of the electron is known as the _spin_ of the electron, which can be thought of as a vector emanating from the electron in space (when thinking classically).

A single quantum spin is the simplest quantum mechanical system, and therefore provides an avenue for understanding quantum mechanics. It is also an instance of a system that can represent a _qubit_, which is analogous to a single bit in a modern digital computer.

## 1.3 An Experiment
The _spin_ can be in one of two states. It can be either in:

$$
\sigma = +1
$$

or it can be in

$$
\sigma = -1.
$$

However, in quantum mechanics we must also consider _measurements_ in addition to the set of states. For now, we will consider our measurement apparatus to be a simple black box that simply reports either $\sigma = +1$ or $\sigma = -1$.

Additionally however, we could flip the apparatus upside down. In this case, a state that was previously measured to be $\sigma = +1$ would be measured as $\sigma = -1$, while the state $\sigma = -1$ would then be measured as $\sigma = +1$.

If we rotate the apparatus by 90 degrees however, we will instead observe the state $\sigma = +1$ as $\sigma = +1$ 50% of the time, and $\sigma = -1$ the remaining 50% of the time. Interestingly, we will observe the same result for the state $\sigma = -1$. If we had thought to treat spin as a classical vector, we can see that the expected rotating the apparatus by 90 degrees should be $\sigma = 0$. However, in quantum mechanics, we see that only the _expected value_ agrees with the classical value, i.e $\langle \sigma \rangle = 0$. This holds true for any angle of rotation, not just 90 degrees.

## 1.4 Experiments Are Never Gentle
If we measure $\sigma = -1$ after having rotated the device, and then we remeasure the spin with the device in it's original orientation, we will now observe 50% probability of $\sigma = +1$ and 50% probability of $\sigma = -1$, _regardless of the original orientation of the spin!_. That means that by measuring the state using the rotated configuration, we prevented ourselves from measuring which state was present in the unrotated configuration of the apparatus. This means that the spin is cannot be in a definite state with respect to the vertical axis and in a definite state with respect to a horizontal at the same time.
## 1.5 Propositions
Classical propositions are constructed from atomic statements, as well as the operations **and** and **or**. For example, we can construct statements such as "Albert Einstein discovered America **or** Isaac Newton was Russion** (false), or "Albert Einstein discovered relativity **and** Isaac Newton was English" (true).
## 1.6 Testing Classical Propositions
To test a classical compound proposition, such as "The die shows an odd-numbered face **and** the die shows a number less than 4", we can test each of the two compoment propositions in either order, and always get the same result.
## 1.7 Testing Quantum Propositions
To test a _quantum_ proposition about the state of of a quantum system, we _cannot_ test component propositions in either order. This means that the underlying proposition is actually meaningless! For example, in the example quantum system described above, the proposition "The spin along the vertical axis is $\sigma = +1$ **and** the spin along the horizontal axis is $\sigma = -1$ is meaningless, since the result will depend on the order in which you test the propositions.
## 1.8 Mathematical Interlude: Complex Numbers
Complex numbers are numbers that can be written as

$$
z = x + yi
$$

where

$$
i = \sqrt{-1}
$$.

We can add, subtract, multiply, and divide them (they form a _field_). Additionally, we can represent them in _polar_ notation, $z = re^{i\theta}$ where $r = \sqrt{x^2 + y^2}$ is the distance from the origin, and $\theta$ is the angle of rotation between the number and the real axis. To multiply complex numbers in polar form, we multiply their distances and add their angles:

$$
z_1z_2 = r_1r_2e^{i(\theta_1 + \theta_2)}
$$

Additionally, each complex number $z = x + yi$ has a complex conjugate $z^{*} = x - yi$. Multiplying a number and it's complex conjugate gives the square of the radius (or the _modulus_).

$$
z^*z = r^2
$$

Note that there is an entire circle's worth of complex numbers with modulus $r = 1$, given by

$$
z = e^{i\theta}.
$$
## 1.9 Mathematical Interlude: Vector Spaces
### 1.9.1 Axioms
1. The sum of any two ket-vectors is a ket-vector
   $$
   | A \rangle + | B \rangle = | C \rangle
   $$
2. Vector addition is commutative:
   $$
   | A \rangle + | B \rangle = | B \rangle + | A \rangle
   $$
3. Vector addition is associative:
   $$
   \big(|A\rangle + |B\rangle\big) + |C\rangle = |A\rangle + \big(|B\rangle + |C\rangle\big)
   $$
4. There is a unique vector $\mathbf{0}$ that is the identity with respect to addition.
   $$
   |A\rangle + \mathbf{0} = |A\rangle
   $$
5. There is a negative of every ket $|A\rangle$, denoted $-|A\rangle$
6. There is a complex scalar multiple of every ket $z|A\rangle$ for every complex number $z$
7. You can distribute scalar multiplication over vector addition and complex addition
   $$
   z\big(|A\rangle + |B\rangle\big) = z|A\rangle + z|B\rangle \\
   \big(z + w\big)|A\rangle = z|A\rangle + w|A\rangle
   $$

### 1.9.2 Functions and Column Vectors
Examples of complex vector spaces:
1. Contiuous complex-valued functions of a single variable $x$
2. Two-dimensional column vectors.

### 1.9.3 Bras and Kets
For every ket vector $|A\rangle$ there is a _dual_ corresponding _bra_ vector $\langle A |$. The terminology derives from the fact that multiplying a _bra_ with a _ket_ results in a _bracket_ (inner product) $\langle A | B \rangle$. The bra of a sum is the sum of the bras, and the bra corresponding to $z|A\rangle$ is $\langle A | z^*$. Bras are analagous to row vectors.
### 1.9.4 Inner Products
Inner products are linear, and the bracket can be interchanged by adding complex conjugation:

$$
\langle B | A \rangle = \langle A | B \rangle^*
$$

### 1.9.5 Orthonormal Bases
Just as in three dimensional space, we can represent any vector as a sum of _basis_ vectors $|i\rangle$

$$
|A\rangle = \sum_{i}{a_i | i \rangle}
$$

Additionally, we have

$$
\langle j | A \rangle = \sum_i{a_i \langle j | i \rangle} = \alpha_j
$$

So

$$
|A\rangle = \sum_i{|i\rangle\langle i | A \rangle}
$$
